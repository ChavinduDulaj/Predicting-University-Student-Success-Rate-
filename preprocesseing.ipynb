Outlier detection and removing

```python
numerical_columns = ['age', 'Medu', 'Fedu', 'traveltime', 'studytime', 'failures', 'famrel', 
                     'freetime', 'goout', 'Dalc', 'Walc', 'health', 'absences', 'G1', 'G2', 'G3']

for column in numerical_columns:
    plt.figure(figsize=(18, 6))
    
    plt.subplot(1, 3, 1)
    sns.boxplot(x=df[column])
    plt.title(f'Boxplot of {column}')
    
    plt.subplot(1, 3, 2)
    sns.histplot(df[column], kde=True, bins=30)
    plt.title(f'Histogram of {column}')
    
    plt.subplot(1, 3, 3)
    sns.violinplot(x=df[column], inner='quartile')
    plt.title(f'Violinplot of {column}')
    
    plt.tight_layout()
    plt.show()
```

```python
# removing the outliers of absences
```

```python
# Calculate mean of 'absences' values between 40 and 60
mean_val = df[(df['absences'] >= 40) & (df['absences'] <= 60)]['absences'].mean()

# Replace values between 40 and 60 in 'absences' column with their mean
df['absences'] = df['absences'].apply(lambda x: mean_val if 40 <= x <= 60 else x)
print(len(df['absences']<mean_val))
# Remove outliers greater than the mean value
df = df[df['absences'] <= mean_val]

```

```python
plt.figure(figsize=(18, 6))
    
plt.subplot(1, 3, 1)
sns.boxplot(x=df['absences'])
plt.title('Boxplot of absences')
    
plt.subplot(1, 3, 2)
sns.histplot(df['absences'], kde=True, bins=30)
plt.title('Histogram of absences')
    
plt.subplot(1, 3, 3)
sns.violinplot(x=df['absences'], inner='quartile')
plt.title('Violinplot of absences')
    
plt.tight_layout()
plt.show()
```

```python
# removing the outliers of G3
```

```python
import random
def modify_list(data):
  """Modifies a list of numbers by replacing zeros in one of three groups with 2.5.

  Args:
      data: A list of numbers.

  Returns:
      A new list with zeros modified as specified.
  """
  group1, group2, group3 = [], [], []
  for num in data:
    if num == 0:
      group1.append(random(0,2.6))  # Replace zero with 2.5 in group 1
    elif len(group2) < len(data) // 3:  # Distribute remaining zeros to group2 up to 1/3 of data length
      group2.append(num)
    else:
      group3.append(num)
  return group1 + group2 + group3

# Example usage
data = df['G3']
modified_data = modify_list(data)
df['G3'] = modified_data
df
```

```python
plt.figure(figsize=(18, 6))
    
plt.subplot(1, 3, 1)
sns.boxplot(x=df['G3'])
plt.title('Boxplot of G3')
    
plt.subplot(1, 3, 2)
sns.histplot(df['G3'], kde=True, bins=30)
plt.title('Histogram of G3')
    
plt.subplot(1, 3, 3)
sns.violinplot(x=df['G3'], inner='quartile')
plt.title('Violinplot of G3')
    
plt.tight_layout()
plt.show()
```
